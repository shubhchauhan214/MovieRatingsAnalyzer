```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

# Movie Ratings Analyzer

```python
np.random.seed(42)

num_movies = 1000

# 1. Movie IDs
movie_ids = np.arange(1, num_movies + 1)

# 2. Genres
genres_list = np.array(['Action', 'Drama', 'Comedy', 'Horror', 'Romance', 'Sci-Fi'])
genres = np.random.choice(genres_list, size=num_movies)

# 3. Release Years
release_years = np.random.randint(2000, 2026, size=num_movies)

# 4. Duration (minutes)
durations = np.random.randint(60, 181, size=num_movies)

# 5. User Ratings (1.0 to 5.0)
user_ratings = np.round(np.random.uniform(1.0, 5.0, size=num_movies), 1)
```

Data Exploration and Analysis Using NumPy
1. Average, Minimum, Maximum and Median Rating
2. Mose Common Genre
3. Year-wise average Rating
4. Genre-wise Average Rating

```python
print("Average Rating:", np.mean(user_ratings))
print("Minimum Rating:", np.min(user_ratings))
print("Maximum Rating:", np.max(user_ratings))
print("Median Rating:", np.median(user_ratings))
```

```python
# Most common Genre
unique_genres, counts = np.unique(genres, return_counts=True)
most_common_genre = unique_genres[np.argmax(counts)]

print("Most Common Genre:", most_common_genre)
```

```python
# Year-wise Average Ratings
years = np.unique(release_years)
yearly_avg_ratings = [np.mean(user_ratings[release_years == year]) for year in years]

for i in range(len(years)):
    print(f"{years[i]}: {yearly_avg_ratings[i]: .2f}")
```

```python
# Genre-wise Average Ratings
genre_avg_ratings = {genre: np.mean(user_ratings[genres == genre]) for genre in genres_list}

for genre, avg in genre_avg_ratings.items():
    print(f"{genre}: {avg:.2f}")
```

Data Visualization using Matplotlib
1. Genre-wise Average Rating (Bar Chart)
2. Year-wise Average Rating (line Chart)
3. Distribution of Ratings (Histogram)
4. Movie Durations Summary (Boxplot)

```python
# Bar Plot - Genre-wise Average Rating
plt.figure(figsize=(8,5))
plt.bar(genre_avg_ratings.keys(), genre_avg_ratings.values(), color='skyblue', zorder=2)
plt.xlabel("Genres")
plt.ylabel("Average User Rating")
plt.title("Average Rating By Genre")
plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=1)
plt.tight_layout()
plt.show()
```

```python
# Line Chart - Year-wise Average Ratings
plt.figure(figsize=(10,5))
plt.plot(years, yearly_avg_ratings, marker='o', color='green')
plt.xlabel("Release Year")
plt.ylabel("Average Rating")
plt.title("Year-wise Average Ratings")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Histogram - Ratings Distribution
plt.figure(figsize=(8,5))
plt.hist(user_ratings, bins=20, color='orange', edgecolor='black', zorder=2)
plt.xlabel("User Ratings")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Ratings")
plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=1)
plt.tight_layout()
plt.show()
```

```python
# Boxplot - Movie Duration Summary
plt.figure(figsize=(5,6))
plt.boxplot(durations, vert=True)
plt.ylabel("Duration (minutes)")
plt.title("Boxplot of Movie Durations")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
